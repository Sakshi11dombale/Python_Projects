## Density Plot

```python
plt.subplot(3,2,1)
x=sns.kdeplot(hd['age'])

plt.subplot(3,2,2)
x=sns.kdeplot(hd['sex'])

plt.subplot(3,2,3)
x=sns.kdeplot(hd['fbs'])

plt.subplot(3,2,4)
x=sns.kdeplot(hd['cp'])

plt.subplot(3,2,5)
x=sns.kdeplot(hd['chol'])

plt.subplot(3,2,6)
x=sns.kdeplot(hd['thal'])
plt.show()
```
