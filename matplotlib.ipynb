## Pie Chart

```python
x=([23,34,123,4])
plt.pie(x,labels=['a','b','c','d'],autopct='%3.2f%%',colors=['#003f5c','#0d7576','#2c9461','#8ca829'])
plt.legend(loc='lower right')
plt.show()
```

```python
x=([23,34,123,4])
A=['a','b','c','d']
C=['#003f5c','#0d7576','#2c9461','#8ca829']
plt.pie(x,labels=A,autopct='%3.2f%%',colors=C)
plt.legend(loc='lower right')
plt.show()
```

```python
x=([23,34,123,4])
A=['a','b','c','d']
C=['#003f5c','#0d7576','#2c9461','#8ca829']
B=[0,0,0,0.50]
plt.pie(x,labels=A,autopct='%3.2f%%',colors=C,explode=B)
plt.legend(loc='lower right')
plt.show()
```

## Bar Chart

```python
x=([23,34,123,4])
A=['a','b','c','d']
C=['#003f5c','#0d7576','#2c9461','#8ca829']
#B=[0,0,0,0.50]
plt.bar(A,x,color=C)
#ax.legend(title='random')
plt.legend(loc='lower right')
plt.show()
```

## Horizontal Bar Chart

```python
x=([23,34,123,4])
A=['a','b','c','d']
C=['#003f5c','#0d7576','#2c9461','#8ca829']
#B=[0,0,0,0.50]
plt.barh(A,x,color=C)
#ax.legend(title='random')
plt.legend(loc='lower right')
plt.show()
```

```python
x=([23,34,123,4])
A=['a','b','c','d']
C=['#003f5c','#0d7576','#2c9461','#8ca829']
#B=[0,0,0,0.50]
plt.bar(A,x,color=C,label=A)
plt.xlabel("AA")
plt.ylabel("BB")
plt.legend(loc='upper left')
plt.show()
```
